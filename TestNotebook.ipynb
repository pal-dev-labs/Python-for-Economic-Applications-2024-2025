```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Create a DataFrame
vector = np.array([1,2,3,4])
df = pd.DataFrame({'Fruit': ['Apple', 'Banana', 'Orange'], 'Count': [10, 5, 3]})

# Plot a pie chart of the fruit counts
df['Count'].plot.pie(autopct="%1.1f%%", startangle=90)

# Add a title and labels
plt.title("Pie Chart of Fruit Counts")
plt.ylabel("Fruit")

# Show the plot
plt.show()
```
